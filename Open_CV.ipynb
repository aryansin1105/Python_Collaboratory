```python
import cv2 #importing opencv
from google.colab.patches import cv2_imshow
#colorful image
image = cv2.imread("/content/opencvTutorial/images/color.jpg") #reading image from files
cv2_imshow(image)
```

```python
print(image.shape) #no of rows and columns and no of channels available on image
```

```python
#grayscale image (black and white image)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
print(gray.shape) #no channels available in gray image
```

```python
#HSV Image
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
print(hsv.shape) #HSV - Hue Saturation Variance
cv2_imshow(hsv)
```

Edge Detection

```python
canny_image = cv2.Canny(image,150,200) #canny edge detector
cv2_imshow(canny_image)
print(canny_image.shape)
```

```python
import numpy as np
#Erosion and Dilation
kernel = np.ones((1,1),np.uint8)
erode_image = cv2.erode(canny_image,kernel,iterations=1)
cv2_imshow(erode_image)
print(erode_image.shape)
```

```python
#Dilation
kernel = np.ones((5,5),np.uint8)
dilate_image = cv2.dilate(image,kernel,iterations=1)
cv2_imshow(dilate_image)
```
